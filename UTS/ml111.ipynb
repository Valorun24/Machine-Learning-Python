```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
import pandas as pd
latih = pd.read_csv("train.csv")
tes = pd.read_csv("test.csv")
latih
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
kategori=['Gender','Customer Type', 'Type of Travel', 'Class']
encoding = pd.DataFrame(
    ohe.fit_transform(latih[kategori]).toarray(),
    columns=['Female', 'Male', 'disloyal', 'Loyal', 'Business', 'Personal', 'Business', 'Eco', 'Eco Plus']
)
encoding
fitur = pd.concat([encoding, latih.drop(kategori, axis=1)], axis=1)
#  print(fitur)

x = fitur[fitur.columns[:-1]]
y = fitur[fitur.columns[-1]]
# print(x)
print(y)
# mengganti value
x.fillna(x.mean(), inplace=True)
fitur_terbaik = SelectKBest(score_func=f_classif, k=7)
feat = fitur_terbaik.fit(x, y)


# mendapatkan array
msk = feat.get_support()
print(msk)

nama_fitur = []
for bool, feature in zip(msk, latih.columns):
    if bool:
        nama_fitur.append(feature)
print(nama_fitur)
```

```python
fiture_data_train = latih[nama_fitur]
target_data_train = latih['Satisfaction']

fiture_data_test = tes[nama_fitur]
target_data_test = tes['Satisfaction']
```

```python
#mengubah data frame
endcode_train = pd.DataFrame(
    ohe.fit_transform(fiture_data_train[['Class']]).toarray(),
    columns=['Business_Class', 'Eco_Class', 'Eco_Plus_Class'])

endcode_test = pd.DataFrame(ohe.fit_transform(fiture_data_test[['Class']]).toarray(),
    columns=['Business_Class', 'Eco_Class', 'Eco_Plus_Class'])

# menggabungkan data kembali
train_finis = pd.concat([endcode_train, fiture_data_train.drop(['Class'], axis=1)], axis=1)
test_finis = pd.concat([endcode_test, fiture_data_test.drop(['Class'], axis=1)], axis=1)

# mengisi missing value
endcode_train.fillna(fiture_data_train.mean(), inplace=True)
endcode_test.fillna(fiture_data_test.mean(), inplace=True)

print(train_finis)
print(test_finis)
```

```python
# Mencari akurasi tertinggi nilai K dengan metode KNN
from sklearn.neighbors import KNeighborsClassifier

# Menyimpan data feature dan datarget variabele
fitur_test = test_finis
target_test = target_data_test
fitur_train = train_finis
target_train = target_data_train

for i in [3, 5, 7, 9, 11]:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(fitur_train, target_train)
    akurasi = knn.score(fitur_test, target_test)
    print(f'Acurasi KNN dengan k:{i}= {akurasi}')
```
