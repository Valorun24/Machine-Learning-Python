```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
import pandas as pd
data_train = pd.read_csv("train.csv")
data_test = pd.read_csv("test.csv")
print(data_train)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
category =['Gender','Customer Type', 'Type of Travel', 'Class']
encoding = pd.DataFrame(
    ohe.fit_transform(data_train[category]).toarray(),
    columns=['Female', 'Male', 'disloyal', 'Loyal', 'Business', 'Personal', 'Business', 'Eco', 'Eco Plus']
)
encoding
feature = pd.concat([encoding, data_train.drop(category, axis=1)], axis=1)

x = feature[feature.columns[:-1]]
y = feature[feature.columns[-1]]

# print(x)
print(y)

# menggantikan value
x.fillna(x.mean(), inplace=True)
best_feature = SelectKBest(score_func=f_classif, k=7)
fit = best_feature.fit(x, y)


# Mendapatkan array boolean dari fitur yang didapat
mask = fit.get_support()
print(mask)

# Mendapatkan nama-nama fitur
feature_select = []
for bool, feature in zip(mask, data_train.columns):
    if bool:
        feature_select.append(feature)
print(feature_select)
```

```python
feature_train = data_train[feature_select]
target_train = data_train['Satisfaction']

feature_test = data_test[feature_select]
target_test = data_test['Satisfaction']

print("\nFeature matrix : \n", x)
print("\nResponse vector : \n", y)
```

```python
#mengubah menjadi dataframe
encode_data_train = pd.DataFrame(
    ohe.fit_transform(feature_train[['Class']]).toarray(),
    columns=['Business_Class', 'Eco_Class', 'Eco_Plus_Class'])

encode_data_test = pd.DataFrame(ohe.fit_transform(feature_test[['Class']]).toarray(),
    columns=['Business_Class', 'Eco_Class', 'Eco_Plus_Class'])

# menggabungkan data kembali
train_final= pd.concat([encode_data_train, feature_train.drop(['Class'], axis=1)], axis=1)
test_final = pd.concat([encode_data_test, feature_test.drop(['Class'], axis=1)], axis=1)

# mengisi missing value
encode_data_train.fillna(feature_train.mean(), inplace=True)
encode_data_test.fillna(feature_test.mean(), inplace=True)

print(train_final)
print(test_final)
```

```python
# Mencari akurasi tertinggi nilai K dengan metode KNN
from sklearn.neighbors import KNeighborsClassifier

# Menyimpan data feature dan data target dalam variabel
feature_test = test_final
target_test = target_test
feature_train = train_final
target_train = target_train

for i in [3, 5, 7, 9, 11]:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(feature_train, target_train)
    accuracy = knn.score(feature_test, target_test)
    print(f'Akurasi KNN dengan k:{i}= {accuracy}')
```
